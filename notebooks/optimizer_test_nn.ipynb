# Testing optimizers Adam vs SDG
we compare Adam vs SDG to compare MSE loss, computation time to choose for main training

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "../results/optimizer_comparison.csv"
df = pd.read_csv(csv_path)
#split by optimizer
sgd  = df[df["optimizer"] == "sgd"]
adam = df[df["optimizer"] == "adam"]
plt.figure(figsize=(6, 4))
plt.plot(sgd["epoch"],  sgd["train_mse"], label="sgd train")
plt.plot(adam["epoch"], adam["train_mse"], label="adam train")
plt.xlabel("epoch")
plt.ylabel("train mse log")
plt.yscale("log")          
plt.title("training mse: sgd vs adam N=10000")
plt.legend()
plt.tight_layout()
plt.show()


```

# Testing different NN architures
After chosing the architecture with the lowest MSE, now to demonstrate the affect, graph training vs test mse for lowest minizing vs highest minimizing


```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "../results/low_vs_high.csv"
df = pd.read_csv(csv_path)
low  = df[df["arch"] == "low"]
high = df[df["arch"] == "high"]
plt.figure(figsize=(10, 4))
#left: low mse vs epoch
plt.subplot(1, 2, 1)
plt.plot(low["epoch"], low["train_mse"], label="low train")
plt.plot(low["epoch"], low["test_mse"],  label="low test")
plt.xlabel("epoch")
plt.ylabel("mse log")
plt.yscale("log")           
plt.title("4 hidden layers, 256 neurons architecture")
plt.legend()
#right: high mse vs epoch
plt.subplot(1, 2, 2)
plt.plot(high["epoch"], high["train_mse"], label="high train")
plt.plot(high["epoch"], high["test_mse"],  label="high test")
plt.xlabel("epoch")
plt.ylabel("mse log")
plt.yscale("log")           
plt.title("2 hidden layers, 32 neurons architecture")
plt.legend()
plt.tight_layout()
plt.show()
```

# Samples vs Training Test
Examining sample size vs test and training MSE

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "../results/optimal_arch_datasets.csv"
df = pd.read_csv(csv_path)
# distinct dataset sizes
sizes = sorted(df["dataset_size"].unique())
plt.figure(figsize=(10, 4))
#left: train mse vs epoch 
plt.subplot(1, 2, 1)
for n in sizes:
    sub = df[df["dataset_size"] == n]
    plt.plot(sub["epoch"], sub["train_mse"], label=f"N={n}")
plt.xlabel("epoch")
plt.ylabel("train mse log")
plt.yscale("log")          # optional but nice
plt.title("training mse vs epoch")
plt.grid(True)
plt.legend()
#right: test mse vs epoch 
plt.subplot(1, 2, 2)
for n in sizes:
    sub = df[df["dataset_size"] == n]
    plt.plot(sub["epoch"], sub["test_mse"], label=f"N={n}")
plt.xlabel("epoch")
plt.ylabel("test mse log")
plt.yscale("log")
plt.title("test mse vs epoch")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
