```python
import pandas as pd
import matplotlib.pyplot as plt

conv = pd.read_csv("../results/calibration_lm_convergence_SPX_A.csv")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(conv["eval"], conv["mse"])
plt.yscale("log")
plt.xlabel("evaluation")
plt.ylabel("MSE")
plt.title("LM convergence: MSE")

plt.subplot(1, 2, 2)
plt.plot(conv["eval"], conv["grad_norm"])
plt.yscale("log")
plt.xlabel("evaluation")
plt.ylabel("||g||")
plt.title("LM convergence: gradient norm")
plt.tight_layout()
plt.show()
```
